```python
import numpy as np
import pandas as pd
```

```python
# 상대경로로 CSV 읽기
df = pd.read_csv("mlb_season_data.csv")

# 상위 5행 미리보기
df.head()
```

```python
df.info()
```

```python
df['first_name'] = df['first_name'].fillna('(Unknown)')
```

```python
# 'position'과 'team' 컬럼 제거
df = df.drop(columns=["team", "link"])

# 결과 확인
df.head()
```

```python
# 컬럼 이름 매핑
mlb_to_kbo_columns = {
    "batting_average": "AVG",
    "games_played": "G",
    "at_bats": "AB",
    "runs": "R",
    "hits": "H",
    "doubles": "2B",
    "triples": "3B",
    "homeruns": "HR",
    "rbi": "RBI",
    "stolen_bases": "SB",
    "caught_stealing": "CS",
    "walks": "BB",
    "strikeouts": "SO",
    "slugging_percentage": "SLG",
    "on_base_percentage": "OBP",
    "on_base_plus_slugging" : "OPS",
}

# 컬럼 이름 변경
df.rename(columns=mlb_to_kbo_columns, inplace=True)

# 확인
df.head()
```

```python
df = df[df['position'] != 'P']
```

```python
df.info()
```

```python
# 동일 인물 기준 그룹화
years = df.groupby(['first_name', 'last_name'])['season'].agg(first_year='min', last_year='max').reset_index()

# 원본 df에 병합 (left join)
df = df.merge(years, on=['first_name', 'last_name'], how='left')

# 결과 확인
df.head()
```

```python
df.info()
```

```python
# RBI가 숫자로 변환되지 않는 경우를 NaN으로 바꿔서 필터링
invalid_rbi = df[pd.to_numeric(df['RBI'], errors='coerce').isna()]

# 결과 확인
print(invalid_rbi)
```

```python
target_cols = ['RBI', 'SB', 'CS']
```

```python
# 먼저 대상 컬럼을 숫자로 바꾸되, 에러는 NaN으로 처리
for col in target_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 사람 단위로 그룹화
grouped = df.groupby(['first_name', 'last_name'])

# 각 컬럼의 평균으로 NaN 채우기
for col in target_cols:
    df[col] = grouped[col].transform(lambda x: x.fillna(x.mean()))
```

```python
df.info()
```

```python
# 비율 제거
df = df.drop(columns=["AVG", "OBP", "SLG", "OPS"])

# 결과 확인
df.head()
```

```python
import pandas as pd

# 예시 데이터 불러오기 (df로 가정)

# 1. 그룹 기준
group_cols = ['first_name', 'last_name']

# 2. 합산 대상 컬럼 정의 (position, first_year, last_year 제외)
sum_cols = [col for col in df.columns if col not in group_cols + ['position', 'first_year', 'last_year']]

# 3. 그룹별 합산
sum_df = df.groupby(group_cols)[sum_cols].sum().reset_index()

# 4. active_year 추가 (사람별 행 수 계산)
active_year_df = df.groupby(group_cols).size().reset_index(name='active_year')

# 5. first_year, last_year, position은 대표값 하나씩만 사용
meta_df = df.groupby(group_cols)[['position', 'first_year', 'last_year']].first().reset_index()

# 6. 모두 합치기
result = sum_df.merge(active_year_df, on=group_cols)
result = result.merge(meta_df, on=group_cols)

# 7. 컬럼 정렬 (선택)
ordered_cols = ['first_name', 'last_name', 'position', 'active_year', 'first_year', 'last_year'] + sum_cols
result = result[ordered_cols]

# 결과 출력
result.head()
```

```python
result.info()
```

```python
result['name'] = result['first_name'] + ' ' + result['last_name']

# 컬럼 순서 정리 (name을 맨 앞으로)
cols = ['name'] + [col for col in result.columns if col != 'name']
result = result[cols]
```

```python
result.head(10)
```

```python
result = result.drop(columns=['first_name', 'last_name', 'season'], errors='ignore')
```

```python
result.head(20)
```

```python
result = result[result['AB'] >= 30]
```

```python
result.info()
```

```python
result.head()
```

```python
# 1루타 (1B) 계산
result['1B'] = result['H'] - result['2B'] - result['3B'] - result['HR']

# AVG 계산
result['AVG'] = result['H'] / result['AB']

# OBP 계산 (간략화: HBP, SF 등 없음)
result['OBP'] = (result['H'] + result['BB']) / (result['AB'] + result['BB'])

# SLG 계산
result['SLG'] = (
    result['1B'] + 2 * result['2B'] + 3 * result['3B'] + 4 * result['HR']
) / result['AB']

# OPS 계산
result['OPS'] = result['OBP'] + result['SLG']

# 소수점 정리 (선택)
result[['AVG', 'OBP', 'SLG', 'OPS']] = result[['AVG', 'OBP', 'SLG', 'OPS']].round(3)

# 1B는 계산용이므로 제거 (선택)
result.drop(columns=['1B'], inplace=True)
```
