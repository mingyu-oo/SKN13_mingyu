```python
# ì–‘ë°©í–¥, layer 1ê°œ
lstm2 = nn.LSTM(
	input_size=4,		# feature ìˆ˜
	hidden_size=256,	# unit/node ìˆ˜
	bidirectional=True,
	# num_layers=1 (default)
)

out2, (hidden2, cell2) = lstm2(input_data)
print("out shape : ", out2.shape)		# [seq_len : 30, 100, ìˆœë°©í–¥ + ì—­ë°©í–¥ : 512]
print("hidden shape : ", hidden2.shape)	# [ìˆœë°©í–¥ + ì—­ë°©í–¥ : 2, 100, 256]
print("cell shape : ", cell2.shape)	# [ìˆœë°©í–¥ + ì—­ë°©í–¥ : 2, 100, 256]
```

```python
# ì–‘ë°©í–¥, layer 1ê°œ
lstm3 = nn.LSTM(
	input_size=4,		# feature ìˆ˜
	hidden_size=256,	# unit/node ìˆ˜
	bidirectional=True,
	num_layers=3
)

out3, (hidden3, cell3) = lstm3(input_data)
print("out shape : ", out3.shape)		# [seq_len : 30, 100, ìˆœë°©í–¥ + ì—­ë°©í–¥ : 512]
print("hidden shape : ", hidden3.shape)	# [(ìˆœë°©í–¥ + ì—­ë°©í–¥)*(layer ìˆ˜) : 2*3 = 6 , 100, 256]
print("cell shape : ", cell3.shape)	# [ìˆœë°©í–¥ + ì—­ë°©í–¥ : 2, 100, 256]
```
