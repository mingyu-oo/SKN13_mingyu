```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ğŸŒ¼ ë°ì´í„° ê²½ë¡œ
flowers102_data_dir = "datasets"

# ğŸŒ» Transform ì •ì˜
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

# ğŸ“¦ Flowers102 ë¶ˆëŸ¬ì˜¤ê¸° (train splitë§Œ)
flowers102_trainset = datasets.Flowers102(
    root=flowers102_data_dir,
    download=True,
    split='train',
    transform=transform
)
```

```python
type(flowers102_trainset), isinstance(flowers102_trainset, Dataset)
# Places365
# mnist_trainsetì´ Datasetì„ ìƒì† ë°›ì•„ ë§Œë“¦. => True
```

```python
# ğŸ›’ Dataloader
train_loader = DataLoader(flowers102_trainset, batch_size=8, shuffle=True)
```

```python
# ğŸ” ë°°ì¹˜ í•˜ë‚˜ í™•ì¸
images, labels = next(iter(train_loader))
images, labels
```

```python
# ğŸ¨ ì‹œê°í™” í•¨ìˆ˜
def imshow(imgs, labels):
    imgs = imgs.numpy()
    fig, axs = plt.subplots(1, len(imgs), figsize=(15, 4))
    for i in range(len(imgs)):
        img = np.transpose(imgs[i], (1, 2, 0))  # (C, H, W) â†’ (H, W, C)
        axs[i].imshow(img)
        axs[i].set_title(f"Label: {labels[i].item()}")
        axs[i].axis('off')
    plt.show()

# ğŸ–¼ï¸ ë°°ì¹˜ ì´ë¯¸ì§€ í™•ì¸
imshow(images, labels)
```
