```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 🌼 데이터 경로
flowers102_data_dir = "datasets"

# 🌻 Transform 정의
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

# 📦 Flowers102 불러오기 (train split만)
flowers102_trainset = datasets.Flowers102(
    root=flowers102_data_dir,
    download=True,
    split='train',
    transform=transform
)
```

```python
type(flowers102_trainset), isinstance(flowers102_trainset, Dataset)
# Places365
# mnist_trainset이 Dataset을 상속 받아 만듦. => True
```

```python
flowers102_trainset
```

```python
# 🛒 Dataloader
train_loader = DataLoader(flowers102_trainset, batch_size=8, shuffle=True)
```

```python
# step
len(train_loader)
```

```python
# 🔍 배치 하나 확인
images, labels = next(iter(train_loader))
images, labels
```

```python
# 🎨 시각화 함수
def imshow(imgs, labels):
    imgs = imgs.numpy()
    fig, axs = plt.subplots(1, len(imgs), figsize=(15, 4))
    for i in range(len(imgs)):
        img = np.transpose(imgs[i], (1, 2, 0))  # (C, H, W) → (H, W, C)
        axs[i].imshow(img)
        axs[i].set_title(f"Label: {labels[i].item()}")
        axs[i].axis('off')
    plt.show()

# 🖼️ 배치 이미지 확인
imshow(images, labels)
```
